```python
import datetime

# 1)
number_of_all_visits = 0

# 2)
views_of_visit  = {}

# 3)
user_visit_list, user_visit_number = {}, {}

# 4)
users_in_16_22 = set()

# 5)
max_simultaneus_visits = 0
visit_timestamps, visits_ts_wo_bounce = {}, {}


with open('zadanie_3_dane.txt', 'r') as file:
    file = file.readlines()
    

for line in file:
    if not line.strip():
        continue
    line = line.strip().split('\t')
    user, visit, timestamp = line

# 1)
    number_of_all_visits += 1

# 2)
    views_of_visit.setdefault(visit, 0)
    views_of_visit[visit] += 1

# 3)
    user_visit_list.setdefault(user, [])
    if visit not in user_visit_list[user]:
        user_visit_list[user].append(visit)

# 4)
    date = datetime.datetime.fromtimestamp(int(timestamp))
    time = date.time()
    
    if time >= datetime.time(16) and time <= datetime.time(22):
        users_in_16_22.add(user)

# 5)
    visit_timestamps.setdefault(visit, [])
    
    if timestamp not in visit_timestamps[visit]:
        visit_timestamps[visit].append(timestamp)
        

# 2)
longest_visit = max(views_of_visit.values())

# 3)
for user, visit_list  in user_visit_list.items():
    user_visit_number[user] = len(visit_list)
    
max_user_visit_number = max(user_visit_number.values())

# 4)
sum_users_in_16_22 = len(users_in_16_22)

        
# 5)
for visit in visit_timestamps:
    if len(visit_timestamps[visit]) >= 2:
        visits_ts_wo_bounce[visit] = visit_timestamps[visit]

        
for visit in visits_ts_wo_bounce:
    first_in_list = int(visits_ts_wo_bounce[visit][0])
    last_in_list = int(visits_ts_wo_bounce[visit][-1]) 
    if len(visits_ts_wo_bounce[visit]) >= 2:
        visits_ts_wo_bounce[visit] = set(range(first_in_list, last_in_list + 1))

        
for visit in visits_ts_wo_bounce:
    visit_counter = 0
    main_set = visits_ts_wo_bounce[visit]   
    for another_visit in visits_ts_wo_bounce:
        another_set = visits_ts_wo_bounce[another_visit]  
        if main_set & another_set:
            visit_counter += 1
            main_set = main_set & another_set        
    if visit_counter > max_simultaneus_visits:
        max_simultaneus_visits = visit_counter
                                            


print(f"\nThe number of all visits: {number_of_all_visits}")
print(f"\nThe number of views in longes visit: {longest_visit}")
print(f"\nMaximum number of visits made by user: {max_user_visit_number}")
print(f"\nSum of users between 16 - 22: {sum_users_in_16_22}")
print(f"\nMaximum number of simultaneus visits: {max_simultaneus_visits}") 

```
